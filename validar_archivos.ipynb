```python
import requests

base_file_url = "https://bahianorte.grupocice.com/tmixaprobacionliberacionesws/files"

downloaded_files_content = []

if 'response_data_documentos' in locals() and response_data_documentos and response_data_documentos.get('data'):
    print("Starting file downloads...")
    for doc in response_data_documentos['data']:
        nombre_archivo = doc.get('nombreArchivo')
        nombre_carpeta = doc.get('nombreCarpeta')

        if nombre_archivo and nombre_carpeta and 'token' in locals() and token:
            file_download_url = f"{base_file_url}?filename={nombre_archivo}&filefolder={nombre_carpeta}"
            headers_with_auth = {
                "Authorization": f"Bearer {token}"
            }
            print(f"Attempting to download: {file_download_url}")

            try:
                response_file = requests.get(file_download_url, headers=headers_with_auth)
                response_file.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)

                downloaded_files_content.append({
                    'filename': nombre_archivo,
                    'content': response_file.content # Store raw content (bytes)
                })
                print(f"Successfully downloaded {nombre_archivo}")

            except requests.exceptions.HTTPError as errh:
                print(f"Http Error for {nombre_archivo}: {errh}")
                print(f"Response content: {response_file.text}")
            except requests.exceptions.ConnectionError as errc:
                print(f"Error Connecting for {nombre_archivo}: {errc}")
            except requests.exceptions.Timeout as errt:
                print(f"Timeout Error for {nombre_archivo}: {errt}")
            except requests.exceptions.RequestException as err:
                print(f"OOps: Something Else for {nombre_archivo}: {err}")
            except Exception as e:
                print(f"An unexpected error occurred for {nombre_archivo}: {e}")
        else:
            print(f"Skipping document due to missing info or token: {doc}")

    print(f"Finished downloading files. Total files downloaded: {len(downloaded_files_content)}")
    # You can now process downloaded_files_content list, for example, save them to disk
    # for file_data in downloaded_files_content:
    #     print(f"File '{file_data['filename']}' has content of size {len(file_data['content'])} bytes.")

else:
    print("Error: 'response_data_documentos' not found or empty. Please ensure the previous cell was executed successfully.")
```

```python
from pypdf import PdfReader
import io

def extract_pdf_text(path):
    reader = PdfReader(path)
    text = ""
    for page in reader.pages:
        text += page.extract_text() + "\n"
    return text

extracted_pdf_texts = []

for file_data in downloaded_files_content:
    filename = file_data['filename']
    content = file_data['content']

    # Use io.BytesIO to treat the bytes content as a file
    pdf_file_object = io.BytesIO(content)

    try:
        # Call the existing extract_pdf_text function
        extracted_text = extract_pdf_text(pdf_file_object)
        extracted_pdf_texts.append({
            'filename': filename,
            'text_content': extracted_text
        })
        print(f"Extracted text from {filename}")
    except Exception as e:
        print(f"Error extracting text from {filename}: {e}")

print(f"Finished text extraction. Total documents processed: {len(extracted_pdf_texts)}")
```

```python
EXTRACTION_PROMPT = """
Eres un analista aduanal experto. Tu tarea es:

1. Identificar el TIPO DE DOCUMENTO dentro del PDF.
2. Extraer TODOS los datos relevantes de manera organizada, clara y en formato JSON.
3. El formato de salida debe ser limpio, estándar y sin texto adicional fuera del JSON.
4. Debes identificar SIEMPRE los documentos de transporte, especialmente el BILL OF LADING (BL).

Reglas generales:
- Clasifica cada documento detectado dentro del PDF como:
  * Forma Simplificada del Pedimento
  * Transport Document:
      - BILL OF LADING (BL)
      - MASTER BILL OF LADING (MBL)
      - HOUSE BILL OF LADING (HBL)
      - BOOKING
      - GUÍA MARÍTIMA
      - SHIPPING INSTRUCTION
  * Pedimento simplificado
  * Anexo del pedimento
  * DODA
  * Solicitud de depósito / ingreso
  * Carta Art. 23
  * Documentos digitalizados referenciados

- Extrae todos los campos visibles, sin omitir fechas, números, textos, sellos, contenedores, claves, códigos o notas.
- Mantén valores EXACTOS (sin limpiar, redondear o reformatear).
- Si hay varios documentos en el archivo, devuélvelos así:
  {
    "documentos": [
      { ...doc 1... },
      { ...doc 2... }
    ]
  }

REGLAS ESPECÍFICAS PARA DETECTAR BILL OF LADING (BL):
- Identifica cualquier línea que contenga:
  "Bill of Lading"
  "B/L"
  "BL No"
  "BOL"
  "MBL"
  "HBL"
  "Master B/L"
  "House B/L"
  "Guía marítima"
  "Ocean Bill"
  "Booking"
  "Booking No"
- Un número BL casi siempre tiene formato:
  * 3+ letras + 7–10 dígitos (ej: MEDUFP157201)
  * Puede contener guiones
- Asegúrate de clasificar sí o sí:
  * BL number
  * Carrier (naviera)
  * Vessel
  * Voyage
  * Container(s)
  * Booking reference
  * Shipper, Consignee, Notify Party (si existen)
  * Puerto de carga / descarga

REGLAS PARA FORMA SIMPLIFICADA DEL PEDIMENTO:
- NUM. PEDIMENTO
- T. OPER (IMP|EXP)
- CVE. PEDIMENTO
  * dos dígitos, ejemplo:A1,A3,C1,C2,D1,GC,K1,L1,P1,IN..etc.
- DESTINO
- PESO BRUTO
- ADUANA E/S
- Clave en el RFC
- CURP
- CODIGO DE ACEPTACIÓN
- CLAVE DE LA SECCION ADUANERA DE DESPACHO
- MARCAS, NUMEROS Y TOTAL DE BULTOS
- FECHA DE ENTRADA
- FECHA DE PAGO
- PATENTE
- PEDIMENTO
- ADUANA
- BANCO
- LINEA DE CAPTURA
- IMPORTE DE PAGO
- FECHA DE PAGO
- NUMERO DE OPERACION BANCARIA
- NUMERO DE TRANSACCION SAT
- MEDIO DE PRESENTACIÓN
- MEDIO DE RECEPCIÓN/COBRO
- NUMERO GUIA / ORDEN DE EMBARQUE / BILL OF LADING / BOOKING / ID
- CONTENEDORES / EQUIPO / NUMERO ECONOMICO
- ACUSE DE VALOR
- AGENTE ADUANAL (nombre, razón social, RFC, CURP)
- NUMERO DE SERIE DEL CERTIFICADO
- e.firma
- DESTINO/ORIGEN


REGLAS CRÍTICAS PARA EVITAR MEZCLA ENTRE DOCUMENTOS:

- NUNCA mezcles datos entre documentos diferentes.
- Cada documento debe contener ÚNICAMENTE la información que aparece dentro de sus propios límites textuales.
- Si un contenedor, buque, viaje, sello o peso aparece fuera del bloque de texto del pedimento, NO debe ser incluido dentro del pedimento.
- No intentes “completar” campos faltantes usando datos de otros documentos.
- Si un documento no contiene un campo (por ejemplo contenedores), entonces ese campo debe OMITIRSE y NO debe buscarse en otro documento.
- Cada documento debe mapearse por separado dentro de un arreglo JSON así:
  {
    "documentos": [
      {
        "tipo_documento": "PEDIMENTO",
        "contenido": { ...solo datos del pedimento... }
      },
      {
        "tipo_documento": "BILL OF LADING",
        "contenido": { ...solo datos del BL... }
      },
      {
        "tipo_documento": "SOLICITUD",
        "contenido": { ...solo datos de la solicitud... }
      }
    ]
  }

- Solo asigna un campo a un documento si aparece explícitamente en el mismo bloque textual (mismo párrafo, misma página o sección del PDF).

- SI TIENES DUDA sobre a cuál documento pertenece una información:
  → colócala en un documento separado bajo “documento_no_identificado” en lugar de mezclarla.

- En la Forma Simplificada del Pedimento:
  → SOLO incluya los contenedores si aparecen explícitamente dentro del mismo cuadro/tablero de datos del pedimento.


Tu salida final será SOLO JSON válido.

Texto del documento:
---
{DOCUMENT_TEXT}
---
"""

```

```python
from google.colab import userdata
from openai import OpenAI # Using OpenAI client for ChatGPT/DeepSeek

OPENAI_API_KEY = userdata.get('OPENAI_KEY_VALIDAR_DOCUMENTOS')
DEEPSEEK_API_KEY = userdata.get('DEEPSEEK_KEY_VALIDAR_DOCUMENTOS')

```

```python
USAR_IA = 'DEEPSEEK' # DEEPSEEK | OPENAI
if USAR_IA == 'OPENAI':
    print("Usando OpenAI")

    # Crear el cliente para OpenAI
    client = OpenAI(api_key=OPENAI_API_KEY)

    # función para enviarle el pdf al modelo gpt-4o-mini
    def parse_document(document_text, model="gpt-4o-mini"):
        prompt = EXTRACTION_PROMPT.replace("{DOCUMENT_TEXT}", document_text)

        response = client.chat.completions.create(
            model=model,
            messages=[
                {"role": "user", "content": prompt}
            ]
        )

        return response.choices[0].message.content
else:
    print("Usando DeepSeek")

    # DeepSeek API base URL
    DEEPSEEK_BASE_URL = "https://api.deepseek.com/v1"

    # Initialize OpenAI client configured for DeepSeek
    client = OpenAI(api_key=DEEPSEEK_API_KEY, base_url=DEEPSEEK_BASE_URL)

    # función para enviarle el pdf al modelo deepseek-coder
    def parse_document(document_text, model="deepseek-coder"):
        prompt = EXTRACTION_PROMPT.replace("{DOCUMENT_TEXT}", document_text)

        response = client.chat.completions.create(
            model=model,
            messages=[
                {"role": "user", "content": prompt}
            ]
        )

        return response.choices[0].message.content
```

```python
# Procesa cada PDF extraído con el modelo
processed_documents_ai_output = []

if 'extracted_pdf_texts' in locals() and extracted_pdf_texts:
    for item in extracted_pdf_texts:
        filename = item['filename']
        text_content = item['text_content']
        print(f"\n--- Processing {filename} with AI model ---")
        try:
            ai_output = parse_document(text_content)
            processed_documents_ai_output.append({
                'filename': filename,
                'ai_extracted_data': ai_output
            })
            print("AI Model Output:")
            print(ai_output)
        except Exception as e:
            print(f"Error processing {filename} with AI model: {e}")
            processed_documents_ai_output.append({
                'filename': filename,
                'ai_extracted_data': f"Error: {e}"
            })
    print("\nFinished processing all documents with AI model.")
else:
    print("Error: 'extracted_pdf_texts' not found or empty. Please ensure text extraction was successful.")
```
